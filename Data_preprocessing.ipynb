thumb finger angle

```python
thumb_cmc = []
for index, row in openData.iloc[:,0:6].iterrows():
        #Loop through joint sets 
        a = np.array([row[0], row[1]]) # First coord
        b = np.array([row[2], row[3]]) # Second coord
        c = np.array([row[4], row[5]]) # Third coord
        
        radians = np.arctan2(c[1] - b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
        angle = np.abs(radians*180.0/np.pi)

        if angle > 180.0:
            angle = 360-angle
        thumb_cmc.append(angle)
```
